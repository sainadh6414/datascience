In this particular session, we will look at how to read Time Series data and look at various measures of decomposition of the Time Series data. We will also look at various exploratory data analysis measures for Time Series data.

Case Study - We will analyse the data of monthly beer production of Australia in the  20th century.

## Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
sns.set()
```

## Loading data as a time series object

```python
df = pd.read_csv('Beer Australia.csv', parse_dates=['Month'],
                index_col='Month')
```

Note that the above method is slightly different from the way we load a dataset normally, since here we are dealing with a time series data. We need to tell python this by specifying the column that contains data info and also it is recommended to set the date column as the index(row index) of the dataframe. This will help us to easily slice and dice the dataset later for further analysis.

### Average monthly production across years

```python
df_yearly_mean.head() # column header is wrong..
```

```python
df_yearly_mean.columns = ['Average monthly Production']
```

```python
df_yearly_mean.head()
```

```python
df_yearly_mean.plot();
```

The resampled annual figures have smoothened out the seasonality variations and we are able to see only the year on year trends in production (both annual totals as well as monthly average for each year).

### Daily

```python
df_daily_sum = df.resample('D').sum()
df_daily_sum.head()
```

Note, that we only have monthly data for each year and we are asking python to give us daily data for each month. Since this is not possible, we are getting the monthly figure for that month assigned to a particular day and for the rest of the days in the month, the value assigned is zero. This is the default behaviour. We should not try to analyze at a granularity finer than the frequency at which we have the time series values.

```python
# plotting this is not going to be very useful
df_daily_sum.plot();
```

### Decade

```python
df_decade_sum = df.resample('10Y').sum()
df_decade_sum.columns = ['Decade Beer Production']
df_decade_sum.head()
```

```python
df_decade_sum.plot();
```

## Time Series Decomposition

### Additive Model

```python
add_decom = seasonal_decompose(df, model='additive')
add_decom.plot();
```

As per the 'additive' decomposition, we see that there is a pronounced trend in the earlier years of the data. There is a seasonality as well.

```python
trend = add_decom.trend
trend.head(12)
```

```python
add_decom.trend[465:]
```

```python
seasonality = add_decom.seasonal
seasonality[:12]
```

```python
seasonality.min(), seasonality.max()
```

```python
residual = add_decom.resid
residual.head(12)
```

```python
residual.mean()
```

```python
# resid + seasonality+trend =Total value of time series
-0.562403-15.491764+86.154167
```

```python
df.head(12)
```

### De-Seasonalized Time Series

```python
deseasonalized_ts = trend + residual
deseasonalized_ts[:20]
```

```python
df.plot(color='black')
deseasonalized_ts.plot(color='red');
plt.legend(['Original Time Series', 'Deseasoanlized Time Series']);
```

```python
mult_decom =seasonal_decompose(df, model = 'multiplicative')
mult_decom.plot();
```

## Missing Value Handling in Time Series

Methods of imputation for replacing missing values (meaningful values)

| Method | When suitable |
| ---------------------------- | ------------------------------------ |
| Take average of the nearest neighbours | Data has no seasonality |
| Take average of the seasons from two or all available years | Data has seasonality |
| Interpolate function of pandas |  |
| Linear interpolation | Relationship in the interval of two samples is a first order polynomial |
| Polynomial such as Quadratic or Cubic interpolation | Second or third order polynomial describes the interval between two samples |
| Spline | Handles non-uniform spacing of samples |

Since this data does not contain any missing values, we will 'create' a missing value and then impute it using the various methods available.

```python
df.isnull().sum().sum()
```

```python
df.loc['1979'].tail()
```

```python
df_miss = df.replace(to_replace=183.1,value=np.nan)
```

```python
df_miss.loc['1979'].tail()
```

```python
df_miss.isnull().sum()
```

### Interpolate

```python
df_miss.interpolate().loc['1979'].tail()
```

```python
df_miss.interpolate(method='spline', order=1).loc['1979'].tail()
```

```python
df_miss.interpolate(method='polynomial', order=2).loc['1979'].tail()
```

```python
df_miss.interpolate(method='pad').loc['1979'].tail()
```

For better understanding of the interpolate function for imputing the missing values, please refer to the documentation <a href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.interpolate.html" style="color:red">click here</a>
