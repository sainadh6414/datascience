### Data Preprocessing

```python
df.columns = ['ID', 'LIMIT_BAL', 'GENDER', 'EDUCATION', 'MARRIAGE', 'AGE', 
              'PAY_SEP', 'PAY_AUG', 'PAY_JUL', 'PAY_JUN', 'PAY_MAY', 'PAY_APR',
             'BILL_AMT_SEP', 'BILL_AMT_AUG', 'BILL_AMT_JUL', 'BILL_AMT_JUN', 'BILL_AMT_MAY', 'BILL_AMT_APR',
             'PAY_AMT_SEP', 'PAY_AMT_AUG', 'PAY_AMT_JUL', 'PAY_AMT_JUN', 'PAY_AMT_MAY', 'PAY_AMT_APR',
             'DEFAULT']
```

```python
df.head()
```

```python
df.describe().T
```

```python
# Converting the categorical column into object type

df['GENDER'] = df['GENDER'].astype(object)
df['EDUCATION'] = df['EDUCATION'].astype(object)
df['MARRIAGE'] = df['MARRIAGE'].astype(object)
df['PAY_SEP'] = df['PAY_SEP'].astype(object)
df['PAY_AUG'] = df['PAY_AUG'].astype(object)
df['PAY_JUL'] = df['PAY_JUL'].astype(object)
df['PAY_JUN'] = df['PAY_JUN'].astype(object)
df['PAY_MAY'] = df['PAY_MAY'].astype(object)
df['PAY_APR'] = df['PAY_APR'].astype(object)
df['DEFAULT'] = df['DEFAULT'].astype(object)
```

```python
# Define a function to retrieve count fo levels within each category
def get_level_counts(i):
    print(df[i].value_counts())
    
PAY_X = ['PAY_APR', 'PAY_MAY', 'PAY_JUN', 'PAY_JUL', 'PAY_AUG', 'PAY_SEP']    

for f in PAY_X:
    get_level_counts(f)
    
df['DEFAULT'].value_counts(normalize=True)
```

### Anomalies Check

```python
### Possible case
# Customers who have negative bill amount in last 3 months,
# have not defaulted in last 6 months but are still marked as defaulters

anom1 = df[(df['BILL_AMT_JUL'] < 0) & (df['BILL_AMT_AUG'] < 0) & (df['BILL_AMT_SEP'] < 0) & 
          (df['PAY_APR'] < 1) & (df['PAY_MAY'] < 1) & (df['PAY_JUN'] < 1) & 
          (df['PAY_JUL'] < 1) & (df['PAY_AUG'] < 1) & (df['PAY_SEP'] < 1) &
          (df['DEFAULT'] == 1)]
anom1
```

```python
anom1.describe(include='all').T
```

```python
# Retrieve the index for the anomaly records
list_anom1 = anom1.index
list_anom1
```

```python
df['DEFAULT'].value_counts()
```

```python
df['DEFAULT'] = np.where(df.index.isin(list_anom1), 0, df['DEFAULT'])
df['DEFAULT'].value_counts()
```

### Feature Engineering

```python
# Repayment status columns which have greater value are more prone to become defaulters

df['RISK_VALUE'] = df['PAY_APR'] + df['PAY_MAY'] + df['PAY_JUN'] + df['PAY_JUL'] + df['PAY_AUG'] + df['PAY_SEP']
```

```python
df['RISK_VALUE'].dtype
```

```python
df['RISK_VALUE'].unique()
```

```python
df['RISK_VALUE'] = df['RISK_VALUE'].astype(int)
```

```python
df['AVG_6_MONTHS_BILL'] = (df['BILL_AMT_APR'] + df['BILL_AMT_MAY'] + df['BILL_AMT_JUN'] + df['BILL_AMT_JUL'] + df['BILL_AMT_AUG'] + df['BILL_AMT_SEP'])/6
df['AVG_6_MONTHS_PAY'] = (df['PAY_AMT_APR'] + df['PAY_AMT_MAY'] + df['PAY_AMT_JUN'] + df['PAY_AMT_JUL'] + df['PAY_AMT_AUG'] + df['PAY_AMT_SEP'])/6
```
