## Outlier Treatment

### To check for outliers, we will be plotting the box plots

```python
df.boxplot(column=['INCOME'])
plt.show()
```

```python
df.boxplot(column=['TRAVEL TIME'])
plt.show()
```

```python
df.boxplot(column=['CAR AGE'])
plt.show()
```

```python
df.boxplot(column=['MILES CLOCKED'])
plt.show()
```

```python
def detectOutlier(col):
    sorted(col)
    Q1, Q3 = np.quantile(col, [0.25,0.75])
    IQR = Q3 - Q1
    lower_whisker_val = Q1 - (1.5* IQR)
    upper_whisker_val = Q3 + (1.5* IQR)
    return lower_whisker_val, upper_whisker_val
```

```python
lr, ur = detectOutlier(df['INCOME'])
print(lr, ur)
df['INCOME'] = np.where(df['INCOME'] > ur, ur, df['INCOME'])
df['INCOME'] = np.where(df['INCOME'] < lr, lr, df['INCOME'])

lr, ur = detectOutlier(df['TRAVEL TIME'])
df['TRAVEL TIME'] = np.where(df['TRAVEL TIME'] > ur, ur, df['TRAVEL TIME'])
df['TRAVEL TIME'] = np.where(df['TRAVEL TIME'] < lr, lr, df['TRAVEL TIME'])

lr, ur = detectOutlier(df['MILES CLOCKED'])
df['MILES CLOCKED'] = np.where(df['MILES CLOCKED'] > ur, ur, df['MILES CLOCKED'])
df['MILES CLOCKED'] = np.where(df['MILES CLOCKED'] < lr, lr, df['MILES CLOCKED'])

df.shape
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20,10))
sns.boxplot(x='INCOME', data=df, ax=axes[0][0])
sns.boxplot(x='TRAVEL TIME', data=df, ax=axes[0][1])
sns.boxplot(x='MILES CLOCKED', data=df, ax=axes[1][0])
plt.show()
```
