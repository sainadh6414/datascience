Over the last 4 years, have observed an exponential increase in the credit card payment defaulters. Preliminary Analysis was conducted by the bank to understand if any pattern could be observed based on demographic details so that appropriate checks /rules could be imposed while issuing the credit card.

However, this effort has really not benefited the bank

Hence, they have requested your support. This time bank has also collected some additional information related to their bill amount, payment amount along with how each customer really uses the bank option to carry forward the payment amount to next month or payment delay was observed for how many months or use revolving amount etc.

Your task is two-fold

(i) Identify if bank needs to impose some rules based on the insights generated from historical data so that they can keep an eye on the "potential defaulters"

(ii) Build a model so that the bank can estimate the probability of default.

## Hypothesis Testing AGE and DEFAULT

## T-Test - 2 Sample t-Test Between AGE and Default

- **Null Hypothesis** H0: There is no effect of AGE for being DEFAULTER. mean(AGE) for Defaulters = mean(AGE) for non-Defaulters
- **Alternate Hpoth** H1: There is a significant effect of AGE for being DEFAULTER. mean(AGE) for Defaulters != mean(AGE) for non-Defaulters

```python
group1 = credit[credit['DEFAULT']==1]
group1 =group1['AGE']

group2 = credit[credit['DEFAULT']==0]
group2 =group2['AGE']
```

```python
from scipy.stats import ttest_ind
```

```python
t_statistic, p_value = ttest_ind(group1, group2)
print(t_statistic, p_value)
```

Conclusion pvalue<0.05 therefore reject Null Hypothesis. There is some difference between the age of Defaulters and Non Defaulters.

## Hypothesis Testing AGE and EDUCATION

ANOVA test AGE and EDUCATION
- **Null Hypothesis**, H0: There is no effect of AGE on any category of EDUCATION.
- **Alternate Hpoth** , H1: There is a significant effect of AGE on atleast one category of EDUCATION.

```python
# Calculating t-stats and pvalue

model = ols('AGE~EDUCATION', data = credit).fit()
sms.stats.anova_lm(model)
```

Conclusion
- pvalue<0.05 therefore Reject the Null Hypothesis. There is atleast one category in EDUCATION whose mean age is different from other categories.

```python
# Pairwise Tuckey test
print(pairwise_tukeyhsd(credit['AGE'],credit['EDUCATION']))
```

- 'Others' category has no significant difference from University and Graduate. Otherwise all other categories have significant difference in their AGE.

### Hypothesis Testing PAYSEP and DEFAULT

**T-Test PAYSEP and Default**
- Null Hypothesis, H0: There is no effect of PAYSEP for being DEFAULTER. mean(PAYSEP) for Defaulters = mean(PAYSEP) for non-Defaulters
- Alternate Hpoth, H1: There is a significant effect of PAYSEP for being DEFAULTER. mean(PAYSEP) for Defaulters != mean(PAYSEP) for non-Defaulters

```python
group1 = credit[credit['DEFAULT']==1]
group1 =group1['PAYSEP']

group2 = credit[credit['DEFAULT']==0]
group2 =group2['PAYSEP']
```

```python
t_statistic, p_value = ttest_ind(group1, group2)
print(t_statistic, p_value)
```

**Conclusion**
- pvalue<0.05 therefore Reject the Null Hypothesis. There is significant difference between the average of Defaulters and non defaulters.
